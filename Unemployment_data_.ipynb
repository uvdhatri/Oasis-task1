1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

2. Load Data

```python
# Assuming you have a CSV file containing unemployment data
df = pd.read_csv("Unemployment in India oasis task2.csv")
df
```

3. Explore Data

```python
# Display basic information about the dataset
print(df.info())

# Display the first few rows of the dataset
print(df.head())

# Summary statistics
print(df.describe())
```

4. Data Cleaning

```python
# Check for missing values
print(df.isnull().sum())

# Handle missing values (if any)
df = df.dropna()  # Example: Dropping rows with missing values
```

5. Data Visualization

```python
# Visualize the unemployment rate over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Region', y='Area', data=df)
plt.title('Unemployment Rate Over Time')
plt.xlabel('Region')
plt.ylabel('Area')
plt.show()
```
